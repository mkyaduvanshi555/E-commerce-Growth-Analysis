```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules # helps in to find "customer also bought" feature!
```

```python
df = pd.read_csv('data_UCI.csv',encoding='latin1')
```

```python
print(df.head(10))
```

```python
print('--Finding Negative Quantities--')
print(df[df['Quantity']<0].head())
```

```python
print('\n--Finding Messy Data--')
print(df[df['Description'].str.contains('JUMBO BAG RED',na=False)]['Description'].unique())
```

```python
# Clean the Data, this is the heavy part as it has to clean more than 500,000+ rows!!

df['Description'] = df['Description'].str.strip()

df['InvoiceNo'] = df['InvoiceNo'].astype('str')

df = df[~df['InvoiceNo'].str.contains('C')]

df.dropna(axis=0, subset=['InvoiceNo', 'Description'], inplace=True)
```

```python
print("--Data after cleaning--")
df.head(20)
```

```python
# PIVOT THE DATA (PREPARE THE "BASKET-ANALYSIS")

print("--- Step 5: Pivoting data for 'France'... ---")

basket = (df[df['Country'] == "France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

# Convert all quantities to 1 (bought) or 0 (not bought)
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
# RUN THE ALGORITHM
# This finds the "bought together" patterns
print("--- Step 6: Running the Apriori algorithm... ---")

# A. Find all itemsets that appear in at least 7% of all baskets
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)

# B. Turn those into the final rules (with at least 50% confidence)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
print("\n#####################################################")
print("### TOP 10 RULES ARE:  ###")
print("#####################################################")

# Sort by "lift" (strongest patterns first) and show the top 10
final_rules = rules.sort_values('lift', ascending=False)
print(final_rules.head(10))
```
